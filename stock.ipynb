```
# ============================================================
# 旧Backtester結果の復元・再計算・日本語グラフ再作成
#
# 1. Google Driveから結果を復元
# 2. 不足CSVをtrades.csvから再計算
# 3. 日本語フォントをファイルから直接指定
# 4. グラフ・CSVを再保存
# ============================================================

# ------------------------------------------------------------
# 1. 標準ライブラリ
# ------------------------------------------------------------
import logging
import shutil
import subprocess
from pathlib import Path

import numpy as np
import pandas as pd


# ------------------------------------------------------------
# 2. パス設定
# ------------------------------------------------------------
LOCAL_OUTPUT_DIR = Path(
    "/content/stock/data/bt_out_legacy_direct"
)

DRIVE_ROOT = Path(
    "/content/drive/MyDrive"
)

DRIVE_OUTPUT_DIR = Path(
    "/content/drive/MyDrive/jp_signal/"
    "legacy_backtest_direct"
)

LOCAL_OUTPUT_DIR.mkdir(
    parents=True,
    exist_ok=True,
)


# ------------------------------------------------------------
# 3. Google Driveをマウント
# ------------------------------------------------------------
if not DRIVE_ROOT.exists():
    from google.colab import drive

    drive.mount("/content/drive")
else:
    print(
        "Google Driveはすでに"
        "マウントされています。"
    )

if not DRIVE_ROOT.exists():
    raise FileNotFoundError(
        "Google Driveのマウントに失敗しました。"
    )

DRIVE_OUTPUT_DIR.mkdir(
    parents=True,
    exist_ok=True,
)


# ------------------------------------------------------------
# 4. Google Driveからローカルへ結果を復元
# ------------------------------------------------------------
print("=" * 70)
print("Google Driveからバックテスト結果を復元します")
print("=" * 70)

if DRIVE_OUTPUT_DIR.exists():
    drive_files = [
        path
        for path in DRIVE_OUTPUT_DIR.iterdir()
        if path.is_file()
    ]
else:
    drive_files = []

if drive_files:
    for source_path in drive_files:
        destination_path = (
            LOCAL_OUTPUT_DIR
            / source_path.name
        )

        shutil.copy2(
            source_path,
            destination_path,
        )

        print(
            "復元:",
            source_path.name,
        )
else:
    print(
        "標準保存先にファイルがありません。"
        "\nGoogle Drive内を検索します。"
    )


# ------------------------------------------------------------
# 5. 標準保存先にtrades.csvがなければDrive内を検索
# ------------------------------------------------------------
LOCAL_TRADES_PATH = (
    LOCAL_OUTPUT_DIR
    / "trades.csv"
)

if not LOCAL_TRADES_PATH.exists():
    print("\nGoogle Drive内のtrades.csvを検索します。")

    all_trade_candidates = list(
        DRIVE_ROOT.rglob("trades.csv")
    )

    # legacy_backtestを含むパスを優先
    preferred_candidates = [
        path
        for path in all_trade_candidates
        if (
            "legacy_backtest" in str(path)
            or "bt_out_legacy" in str(path)
        )
    ]

    candidates = (
        preferred_candidates
        if preferred_candidates
        else all_trade_candidates
    )

    if candidates:
        # 更新日時が最も新しいものを採用
        source_trades = max(
            candidates,
            key=lambda path: path.stat().st_mtime,
        )

        source_directory = source_trades.parent

        print("採用する保存先:")
        print(source_directory)

        for source_path in source_directory.iterdir():
            if not source_path.is_file():
                continue

            destination_path = (
                LOCAL_OUTPUT_DIR
                / source_path.name
            )

            shutil.copy2(
                source_path,
                destination_path,
            )

            print(
                "復元:",
                source_path.name,
            )


# ------------------------------------------------------------
# 6. trades.csvの存在確認
# ------------------------------------------------------------
if not LOCAL_TRADES_PATH.exists():
    searched_paths = [
        str(DRIVE_OUTPUT_DIR),
        str(DRIVE_ROOT),
        str(LOCAL_OUTPUT_DIR),
    ]

    raise FileNotFoundError(
        "trades.csvが見つかりません。"
        "\n検索した場所:"
        "\n"
        + "\n".join(searched_paths)
        + "\n\nGoogle Driveにもtrades.csvがない場合は、"
        "旧Backtester本体を再実行する必要があります。"
    )

print("\n使用するtrades.csv:")
print(LOCAL_TRADES_PATH)


# ------------------------------------------------------------
# 7. trades.csv読み込み
# ------------------------------------------------------------
trades = pd.read_csv(
    LOCAL_TRADES_PATH,
    encoding="utf-8-sig",
    dtype={
        "code": str,
    },
)

if trades.empty:
    raise RuntimeError(
        "trades.csvが0件です。"
    )

print("\ntrades.csv:")
print("行数:", f"{len(trades):,}")
print("列:", trades.columns.tolist())


# ------------------------------------------------------------
# 8. 約定取引を抽出
# ------------------------------------------------------------
if "status" in trades.columns:
    filled_trades = trades.loc[
        trades["status"].astype(str)
        == "FILLED"
    ].copy()
else:
    print(
        "注意: status列がないため、"
        "全行を約定取引として扱います。"
    )

    filled_trades = trades.copy()

if filled_trades.empty:
    if "status" in trades.columns:
        print("\nstatus内訳:")
        print(
            trades["status"]
            .value_counts(dropna=False)
            .to_string()
        )

    raise RuntimeError(
        "FILLED取引が0件です。"
    )

if "pnl" not in filled_trades.columns:
    raise KeyError(
        "trades.csvにpnl列がありません。"
        f"\n列: {trades.columns.tolist()}"
    )

filled_trades["pnl"] = pd.to_numeric(
    filled_trades["pnl"],
    errors="coerce",
)

filled_trades = filled_trades.dropna(
    subset=["pnl"]
).copy()

if filled_trades.empty:
    raise RuntimeError(
        "有効なpnlを持つ約定取引が0件です。"
    )

print("\n約定取引:")
print("約定数:", f"{len(filled_trades):,}件")


# ------------------------------------------------------------
# 9. 日付列を整理
# ------------------------------------------------------------
date_columns = [
    "date",
    "entry_date",
    "exit_date",
]

for column in date_columns:
    if column in filled_trades.columns:
        filled_trades[column] = pd.to_datetime(
            filled_trades[column],
            errors="coerce",
        )

if "exit_date" in filled_trades.columns:
    pnl_date_column = "exit_date"
elif "date" in filled_trades.columns:
    pnl_date_column = "date"
else:
    raise KeyError(
        "損益計上に使える日付列がありません。"
    )

filled_trades = filled_trades.dropna(
    subset=[pnl_date_column]
).copy()

if filled_trades.empty:
    raise RuntimeError(
        "有効な決済日を持つ取引が0件です。"
    )


# ------------------------------------------------------------
# 10. 既存の指標CSVから初期資金を取得
# ------------------------------------------------------------
METRICS_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_backtest_metrics.csv"
)

INITIAL_CAPITAL = 100_000_000.0

existing_metrics = pd.DataFrame()

if METRICS_PATH.exists():
    try:
        existing_metrics = pd.read_csv(
            METRICS_PATH,
            encoding="utf-8-sig",
        )

        if {
            "指標",
            "値",
        }.issubset(existing_metrics.columns):

            initial_rows = existing_metrics.loc[
                existing_metrics["指標"].astype(str)
                == "初期資金_円",
                "値",
            ]

            if not initial_rows.empty:
                INITIAL_CAPITAL = float(
                    initial_rows.iloc[0]
                )

    except Exception as exception:
        print(
            "既存指標CSVの読み込みをスキップ:",
            exception,
        )

print("\n初期資金:")
print(f"{INITIAL_CAPITAL:,.0f}円")


# ------------------------------------------------------------
# 11. 取引日カレンダーを復元
# ------------------------------------------------------------
HISTORICAL_SIGNALS_PATH = (
    LOCAL_OUTPUT_DIR
    / "historical_signals.csv"
)

calendar_dates = set()

if HISTORICAL_SIGNALS_PATH.exists():
    try:
        historical_signals = pd.read_csv(
            HISTORICAL_SIGNALS_PATH,
            encoding="utf-8-sig",
            dtype={
                "code": str,
            },
        )

        if "date" in historical_signals.columns:
            signal_calendar = pd.to_datetime(
                historical_signals["date"],
                errors="coerce",
            ).dropna()

            calendar_dates.update(
                signal_calendar.tolist()
            )

    except Exception as exception:
        print(
            "historical_signals.csvの"
            "読み込みをスキップ:",
            exception,
        )

# 決済日は必ずカレンダーに追加
calendar_dates.update(
    filled_trades[
        pnl_date_column
    ].dropna().tolist()
)

# エントリー日も追加
if "entry_date" in filled_trades.columns:
    calendar_dates.update(
        filled_trades[
            "entry_date"
        ].dropna().tolist()
    )

trading_calendar = pd.DatetimeIndex(
    sorted(
        pd.Timestamp(date).normalize()
        for date in calendar_dates
    )
).drop_duplicates()

if trading_calendar.empty:
    raise RuntimeError(
        "バックテスト用取引日を復元できません。"
    )

print("\n復元した取引日:")
print(
    trading_calendar.min().strftime("%Y-%m-%d"),
    "～",
    trading_calendar.max().strftime("%Y-%m-%d"),
)
print(
    "取引日数:",
    len(trading_calendar),
)


# ------------------------------------------------------------
# 12. 日次損益を再計算
# ------------------------------------------------------------
daily_pnl = (
    filled_trades
    .groupby(
        pnl_date_column,
        sort=True,
    )["pnl"]
    .sum()
    .astype(float)
)

daily_pnl.index = pd.DatetimeIndex(
    daily_pnl.index
).normalize()

daily_pnl = (
    daily_pnl
    .groupby(level=0)
    .sum()
    .reindex(
        trading_calendar,
        fill_value=0.0,
    )
    .sort_index()
)

if daily_pnl.empty:
    raise RuntimeError(
        "日次損益を計算できませんでした。"
    )


# ------------------------------------------------------------
# 13. 資産曲線・日次リターンを再計算
# ------------------------------------------------------------
equity_without_initial = (
    INITIAL_CAPITAL
    + daily_pnl.cumsum()
)

initial_date = (
    daily_pnl.index.min()
    - pd.Timedelta(days=1)
)

equity_curve = pd.concat(
    [
        pd.Series(
            [INITIAL_CAPITAL],
            index=[initial_date],
            dtype=float,
        ),
        equity_without_initial,
    ]
)

previous_equity = equity_curve.shift(1)

daily_returns = (
    equity_curve
    / previous_equity
    - 1.0
).iloc[1:]

running_peak = equity_curve.cummax()

drawdown = (
    equity_curve
    / running_peak
    - 1.0
)

drawdown_yen = (
    equity_curve
    - running_peak
)

cumulative_pnl = (
    equity_curve
    - INITIAL_CAPITAL
)


# ------------------------------------------------------------
# 14. 評価指標を再計算
# ------------------------------------------------------------
NUMBER_OF_SIGNALS = int(
    len(trades)
)

NUMBER_OF_FILLED = int(
    len(filled_trades)
)

FILL_RATE = (
    NUMBER_OF_FILLED
    / NUMBER_OF_SIGNALS
    if NUMBER_OF_SIGNALS > 0
    else 0.0
)

TOTAL_PNL = float(
    daily_pnl.sum()
)

FINAL_ASSET = float(
    equity_curve.iloc[-1]
)

TOTAL_RETURN = float(
    FINAL_ASSET
    / INITIAL_CAPITAL
    - 1.0
)

WIN_RATE = float(
    (
        filled_trades["pnl"] > 0
    ).mean()
)

AVERAGE_TRADE_PNL = float(
    filled_trades["pnl"].mean()
)

MEDIAN_TRADE_PNL = float(
    filled_trades["pnl"].median()
)

DAILY_RETURN_MEAN = float(
    daily_returns.mean()
)

DAILY_RETURN_STD = float(
    daily_returns.std(ddof=1)
)

if (
    len(daily_returns) > 1
    and DAILY_RETURN_STD > 0
):
    SHARPE = float(
        DAILY_RETURN_MEAN
        / DAILY_RETURN_STD
        * np.sqrt(252.0)
    )
else:
    SHARPE = 0.0

MAX_DRAWDOWN_PCT = float(
    drawdown.min()
)

MAX_DRAWDOWN_YEN = float(
    drawdown_yen.min()
)

GROSS_PROFIT = float(
    filled_trades.loc[
        filled_trades["pnl"] > 0,
        "pnl",
    ].sum()
)

GROSS_LOSS = float(
    -filled_trades.loc[
        filled_trades["pnl"] < 0,
        "pnl",
    ].sum()
)

if GROSS_LOSS > 0:
    PROFIT_FACTOR = (
        GROSS_PROFIT
        / GROSS_LOSS
    )
elif GROSS_PROFIT > 0:
    PROFIT_FACTOR = np.inf
else:
    PROFIT_FACTOR = np.nan


# ------------------------------------------------------------
# 15. 月次損益を再計算
# ------------------------------------------------------------
monthly_source = pd.DataFrame(
    {
        "date": daily_pnl.index,
        "daily_pnl": daily_pnl.values,
    }
)

monthly_source["month"] = (
    monthly_source["date"]
    .dt.to_period("M")
    .astype(str)
)

monthly_pnl = (
    monthly_source
    .groupby("month")["daily_pnl"]
    .sum()
)


# ------------------------------------------------------------
# 16. 保存用DataFrame
# ------------------------------------------------------------
curve_data = pd.DataFrame(
    {
        "date": equity_curve.index,
        "equity": equity_curve.values,
        "cumulative_pnl": cumulative_pnl.values,
        "running_peak": running_peak.values,
        "drawdown": drawdown.values,
    }
)

daily_data = pd.DataFrame(
    {
        "date": daily_pnl.index,
        "daily_pnl": daily_pnl.values,
        "daily_return": (
            daily_returns
            .reindex(daily_pnl.index)
            .fillna(0.0)
            .values
        ),
    }
)

monthly_data = pd.DataFrame(
    {
        "month": monthly_pnl.index,
        "monthly_pnl": monthly_pnl.values,
    }
)

status_counts = (
    trades["status"]
    .value_counts(dropna=False)
    .rename_axis("status")
    .reset_index(name="count")
    if "status" in trades.columns
    else pd.DataFrame(
        {
            "status": ["UNKNOWN"],
            "count": [len(trades)],
        }
    )
)

metrics = pd.DataFrame(
    {
        "指標": [
            "バックテスト開始日",
            "バックテスト終了日",
            "初期資金_円",
            "最終資産_円",
            "総損益_円",
            "総リターン",
            "日次平均リターン",
            "日次リターン標準偏差",
            "Sharpe比",
            "最大ドローダウン_円",
            "最大ドローダウン率",
            "全シグナル数",
            "約定数",
            "約定率",
            "勝率",
            "平均取引損益_円",
            "中央値取引損益_円",
            "総利益_円",
            "総損失_円",
            "Profit_Factor",
        ],
        "値": [
            daily_pnl.index.min().strftime(
                "%Y-%m-%d"
            ),
            daily_pnl.index.max().strftime(
                "%Y-%m-%d"
            ),
            INITIAL_CAPITAL,
            FINAL_ASSET,
            TOTAL_PNL,
            TOTAL_RETURN,
            DAILY_RETURN_MEAN,
            DAILY_RETURN_STD,
            SHARPE,
            MAX_DRAWDOWN_YEN,
            MAX_DRAWDOWN_PCT,
            NUMBER_OF_SIGNALS,
            NUMBER_OF_FILLED,
            FILL_RATE,
            WIN_RATE,
            AVERAGE_TRADE_PNL,
            MEDIAN_TRADE_PNL,
            GROSS_PROFIT,
            GROSS_LOSS,
            PROFIT_FACTOR,
        ],
    }
)


# ------------------------------------------------------------
# 17. CSVをUTF-8 BOM付きで再保存
# ------------------------------------------------------------
CURVE_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_backtest_curve.csv"
)

DAILY_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_daily_pnl.csv"
)

MONTHLY_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_monthly_pnl.csv"
)

METRICS_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_backtest_metrics.csv"
)

FILLED_TRADES_PATH = (
    LOCAL_OUTPUT_DIR
    / "filled_trades.csv"
)

STATUS_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_status_counts.csv"
)

trades.to_csv(
    LOCAL_TRADES_PATH,
    index=False,
    encoding="utf-8-sig",
)

filled_trades.to_csv(
    FILLED_TRADES_PATH,
    index=False,
    encoding="utf-8-sig",
)

curve_data.to_csv(
    CURVE_PATH,
    index=False,
    encoding="utf-8-sig",
)

daily_data.to_csv(
    DAILY_PATH,
    index=False,
    encoding="utf-8-sig",
)

monthly_data.to_csv(
    MONTHLY_PATH,
    index=False,
    encoding="utf-8-sig",
)

metrics.to_csv(
    METRICS_PATH,
    index=False,
    encoding="utf-8-sig",
)

status_counts.to_csv(
    STATUS_PATH,
    index=False,
    encoding="utf-8-sig",
)

print("\n不足していたCSVを再構築しました。")


# ------------------------------------------------------------
# 18. 評価結果表示
# ------------------------------------------------------------
print("\n" + "=" * 70)
print("旧Backtester 評価結果")
print("=" * 70)
print("初期資金:", f"{INITIAL_CAPITAL:,.0f}円")
print("最終資産:", f"{FINAL_ASSET:,.0f}円")
print("総損益:", f"{TOTAL_PNL:,.0f}円")
print("総リターン:", f"{TOTAL_RETURN:.2%}")
print("Sharpe比:", f"{SHARPE:.3f}")
print(
    "最大ドローダウン:",
    f"{MAX_DRAWDOWN_PCT:.2%}",
    f"({MAX_DRAWDOWN_YEN:,.0f}円)",
)
print("全シグナル:", f"{NUMBER_OF_SIGNALS:,}件")
print("約定数:", f"{NUMBER_OF_FILLED:,}件")
print("約定率:", f"{FILL_RATE:.2%}")
print("勝率:", f"{WIN_RATE:.2%}")
print(
    "平均取引損益:",
    f"{AVERAGE_TRADE_PNL:,.0f}円",
)

if np.isfinite(PROFIT_FACTOR):
    print(
        "Profit Factor:",
        f"{PROFIT_FACTOR:.3f}",
    )
elif np.isinf(PROFIT_FACTOR):
    print("Profit Factor: ∞")
else:
    print("Profit Factor: 算出不可")


# ------------------------------------------------------------
# 19. 日本語フォントを準備
# ------------------------------------------------------------
font_candidates = [
    Path(
        "/usr/share/fonts/opentype/noto/"
        "NotoSansCJK-Regular.ttc"
    ),
    Path(
        "/usr/share/fonts/opentype/noto/"
        "NotoSansCJK-Bold.ttc"
    ),
    Path(
        "/usr/share/fonts/truetype/noto/"
        "NotoSansJP-Regular.ttf"
    ),
]

font_path = next(
    (
        path
        for path in font_candidates
        if path.exists()
    ),
    None,
)

if font_path is None:
    subprocess.run(
        [
            "apt-get",
            "update",
            "-qq",
        ],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        check=True,
    )

    subprocess.run(
        [
            "apt-get",
            "install",
            "-y",
            "-qq",
            "fonts-noto-cjk",
        ],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        check=True,
    )

    font_path = next(
        (
            path
            for path in font_candidates
            if path.exists()
        ),
        None,
    )

if font_path is None:
    raise FileNotFoundError(
        "日本語フォントが見つかりません。"
    )

print("\n使用する日本語フォント:")
print(font_path)


# ------------------------------------------------------------
# 20. Matplotlib設定
# ------------------------------------------------------------
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib.ticker as mticker
import matplotlib.dates as mdates

# グローバルフォントには存在が保証される
# DejaVu Sansを使用する
mpl.rcParams.update(
    {
        "font.family": "DejaVu Sans",
        "axes.unicode_minus": False,
        "figure.dpi": 120,
        "savefig.dpi": 200,
        "figure.facecolor": "white",
        "axes.facecolor": "white",
    }
)

# 日本語はフォントファイルを直接参照する
JP_FONT = fm.FontProperties(
    fname=str(font_path),
)

JP_FONT_TITLE = fm.FontProperties(
    fname=str(font_path),
    size=14,
)

JP_FONT_SUPTITLE = fm.FontProperties(
    fname=str(font_path),
    size=17,
)

JP_FONT_LEGEND = fm.FontProperties(
    fname=str(font_path),
    size=10,
)

resolved_font_path = fm.findfont(
    JP_FONT,
    fallback_to_default=False,
)

print("Matplotlib参照フォント:")
print(resolved_font_path)

# 直接指定したファイルと一致することを確認
if (
    Path(resolved_font_path).resolve()
    != font_path.resolve()
):
    raise RuntimeError(
        "Matplotlibが指定した日本語フォントを"
        "参照していません。"
        f"\n指定: {font_path}"
        f"\n実際: {resolved_font_path}"
    )

logging.getLogger(
    "matplotlib.font_manager"
).setLevel(logging.ERROR)


# ------------------------------------------------------------
# 21. 日本語ラベル設定関数
# ------------------------------------------------------------
def set_japanese_labels(
    axis,
    *,
    title,
    xlabel,
    ylabel,
):
    axis.set_title(
        title,
        fontproperties=JP_FONT_TITLE,
    )

    axis.set_xlabel(
        xlabel,
        fontproperties=JP_FONT,
    )

    axis.set_ylabel(
        ylabel,
        fontproperties=JP_FONT,
    )


# ------------------------------------------------------------
# 22. サマリーグラフ作成
# ------------------------------------------------------------
SUMMARY_GRAPH_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_backtest_summary_japanese.png"
)

fig, axes = plt.subplots(
    2,
    2,
    figsize=(18, 12),
)


# 資産曲線
axes[0, 0].plot(
    curve_data["date"],
    curve_data["equity"],
    color="navy",
    linewidth=1.8,
    label="資産曲線",
)

axes[0, 0].axhline(
    INITIAL_CAPITAL,
    color="gray",
    linestyle="--",
    linewidth=1,
    label="初期資金",
)

set_japanese_labels(
    axes[0, 0],
    title="資産曲線",
    xlabel="日付",
    ylabel="資産額（円）",
)

axes[0, 0].grid(alpha=0.3)

axes[0, 0].legend(
    prop=JP_FONT_LEGEND,
)

axes[0, 0].yaxis.set_major_formatter(
    mticker.StrMethodFormatter(
        "{x:,.0f}"
    )
)


# ドローダウン
axes[0, 1].fill_between(
    curve_data["date"],
    curve_data["drawdown"] * 100,
    0,
    color="crimson",
    alpha=0.45,
)

axes[0, 1].plot(
    curve_data["date"],
    curve_data["drawdown"] * 100,
    color="darkred",
    linewidth=1,
)

set_japanese_labels(
    axes[0, 1],
    title="ドローダウン",
    xlabel="日付",
    ylabel="ドローダウン（%）",
)

axes[0, 1].grid(alpha=0.3)


# 日次損益
daily_colors = np.where(
    daily_data["daily_pnl"] >= 0,
    "steelblue",
    "tomato",
)

axes[1, 0].bar(
    daily_data["date"],
    daily_data["daily_pnl"],
    color=daily_colors,
    width=1.0,
)

axes[1, 0].axhline(
    0,
    color="black",
    linewidth=0.8,
)

set_japanese_labels(
    axes[1, 0],
    title="日次損益",
    xlabel="日付",
    ylabel="日次損益（円）",
)

axes[1, 0].grid(alpha=0.3)

axes[1, 0].yaxis.set_major_formatter(
    mticker.StrMethodFormatter(
        "{x:,.0f}"
    )
)


# 月次損益
monthly_colors = np.where(
    monthly_data["monthly_pnl"] >= 0,
    "steelblue",
    "tomato",
)

axes[1, 1].bar(
    monthly_data["month"].astype(str),
    monthly_data["monthly_pnl"],
    color=monthly_colors,
)

axes[1, 1].axhline(
    0,
    color="black",
    linewidth=0.8,
)

set_japanese_labels(
    axes[1, 1],
    title="月次損益",
    xlabel="年月",
    ylabel="月次損益（円）",
)

axes[1, 1].grid(
    axis="y",
    alpha=0.3,
)

axes[1, 1].tick_params(
    axis="x",
    rotation=45,
)

axes[1, 1].yaxis.set_major_formatter(
    mticker.StrMethodFormatter(
        "{x:,.0f}"
    )
)


# 日付軸
for axis in [
    axes[0, 0],
    axes[0, 1],
    axes[1, 0],
]:
    axis.xaxis.set_major_locator(
        mdates.AutoDateLocator()
    )

    axis.xaxis.set_major_formatter(
        mdates.DateFormatter("%Y-%m")
    )

    axis.tick_params(
        axis="x",
        rotation=30,
    )


fig.suptitle(
    (
        "jp_signal 旧Backtester\n"
        f"総リターン {TOTAL_RETURN:.2%} ／ "
        f"Sharpe比 {SHARPE:.3f} ／ "
        f"最大DD {MAX_DRAWDOWN_PCT:.2%}"
    ),
    fontproperties=JP_FONT_SUPTITLE,
)

plt.tight_layout(
    rect=[
        0,
        0,
        1,
        0.94,
    ]
)

plt.savefig(
    SUMMARY_GRAPH_PATH,
    dpi=200,
    bbox_inches="tight",
    facecolor="white",
)

plt.show()

print("\nサマリーグラフ保存:")
print(SUMMARY_GRAPH_PATH)


# ------------------------------------------------------------
# 23. 損益分布グラフ
# ------------------------------------------------------------
DISTRIBUTION_GRAPH_PATH = (
    LOCAL_OUTPUT_DIR
    / "legacy_trade_pnl_distribution_japanese.png"
)

fig, axis = plt.subplots(
    figsize=(13, 7),
)

axis.hist(
    filled_trades["pnl"],
    bins=30,
    color="steelblue",
    edgecolor="black",
    alpha=0.8,
)

axis.axvline(
    0,
    color="red",
    linestyle="--",
    linewidth=1.5,
    label="損益ゼロ",
)

axis.axvline(
    AVERAGE_TRADE_PNL,
    color="darkgreen",
    linestyle="--",
    linewidth=1.5,
    label=(
        f"平均損益 "
        f"{AVERAGE_TRADE_PNL:,.0f}円"
    ),
)

set_japanese_labels(
    axis,
    title="旧Backtester・取引損益分布",
    xlabel="取引損益（円）",
    ylabel="取引数",
)

axis.grid(alpha=0.3)

axis.legend(
    prop=JP_FONT_LEGEND,
)

axis.xaxis.set_major_formatter(
    mticker.StrMethodFormatter(
        "{x:,.0f}"
    )
)

plt.tight_layout()

plt.savefig(
    DISTRIBUTION_GRAPH_PATH,
    dpi=200,
    bbox_inches="tight",
    facecolor="white",
)

plt.show()

print("\n損益分布グラフ保存:")
print(DISTRIBUTION_GRAPH_PATH)


# ------------------------------------------------------------
# 24. すべての結果をDriveへ再保存
# ------------------------------------------------------------
print("\n" + "=" * 70)
print("Google Driveへ結果を保存します")
print("=" * 70)

for source_path in sorted(
    LOCAL_OUTPUT_DIR.iterdir()
):
    if not source_path.is_file():
        continue

    destination_path = (
        DRIVE_OUTPUT_DIR
        / source_path.name
    )

    shutil.copy2(
        source_path,
        destination_path,
    )

    print(
        "保存:",
        destination_path,
    )


# ------------------------------------------------------------
# 25. 結果表示
# ------------------------------------------------------------
print("\n主要指標:")
from IPython.display import display

display(metrics)

print("\n資産曲線末尾:")
display(curve_data.tail())

print("\n約定取引先頭:")
display(filled_trades.head())


# ------------------------------------------------------------
# 26. 完了表示
# ------------------------------------------------------------
print("\n" + "=" * 70)
print("復元・再計算・日本語グラフ作成が完了しました")
print("=" * 70)

print("\nローカル保存先:")
print(LOCAL_OUTPUT_DIR)

print("\nGoogle Drive保存先:")
print(DRIVE_OUTPUT_DIR)

print("\n主要ファイル:")
print("- trades.csv")
print("- filled_trades.csv")
print("- legacy_backtest_curve.csv")
print("- legacy_daily_pnl.csv")
print("- legacy_monthly_pnl.csv")
print("- legacy_backtest_metrics.csv")
print("- legacy_status_counts.csv")
print("- legacy_backtest_summary_japanese.png")
print("- legacy_trade_pnl_distribution_japanese.png")

print("\nフォント指定方式:")
print(
    "FontProperties(fname=フォントファイル)による"
    "直接指定"
)
```

```
# ============================================================
# バックテスト画像をZIPにまとめてPCへダウンロード
# ============================================================

from pathlib import Path
import shutil
from google.colab import files

# ローカル保存先
LOCAL_OUTPUT_DIR = Path(
    "/content/stock/data/bt_out_legacy_direct"
)

# ダウンロードする画像
image_paths = [
    LOCAL_OUTPUT_DIR
    / "legacy_backtest_summary_japanese.png",

    LOCAL_OUTPUT_DIR
    / "legacy_trade_pnl_distribution_japanese.png",
]

# 存在確認
existing_images = [
    path
    for path in image_paths
    if path.exists()
]

missing_images = [
    path
    for path in image_paths
    if not path.exists()
]

if missing_images:
    print("見つからなかった画像:")

    for path in missing_images:
        print(" -", path)

if not existing_images:
    raise FileNotFoundError(
        "ダウンロードできる画像がありません。"
        "\n先にグラフ作成コードを実行してください。"
    )

# ZIP作成用の一時フォルダ
DOWNLOAD_DIR = Path(
    "/content/legacy_backtest_images"
)

if DOWNLOAD_DIR.exists():
    shutil.rmtree(DOWNLOAD_DIR)

DOWNLOAD_DIR.mkdir(
    parents=True,
    exist_ok=True,
)

# 画像を一時フォルダへコピー
for image_path in existing_images:
    destination = (
        DOWNLOAD_DIR
        / image_path.name
    )

    shutil.copy2(
        image_path,
        destination,
    )

    print(
        "ZIPへ追加:",
        image_path.name,
    )

# ZIPファイルを作成
ZIP_BASE = Path(
    "/content/legacy_backtest_images"
)

ZIP_PATH = Path(
    shutil.make_archive(
        str(ZIP_BASE),
        "zip",
        root_dir=str(DOWNLOAD_DIR),
    )
)

print("\nZIPファイルを作成しました:")
print(ZIP_PATH)
print(
    "ファイルサイズ:",
    f"{ZIP_PATH.stat().st_size / 1024**2:.2f} MB",
)

# PCへダウンロード
files.download(
    str(ZIP_PATH)
)
```
